```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def fix_csv(df):
    df = df.copy()
    df["path"] = df["call_price"]
    df["call_price"] = df.index
    df.set_index("path", drop=True, inplace=True)
    return df
```

```python
antithetic1 = fix_csv(pd.read_csv("antithetic1.csv"))
regular1 = fix_csv(pd.read_csv("regular1.csv"))
antithetic2 = fix_csv(pd.read_csv("antithetic2.csv"))
regular2 = fix_csv(pd.read_csv("regular2.csv"))
```

```python
antithetic1.head()
```

```python
%matplotlib inline
fig = plt.figure(figsize=(6,4))

line0, = plt.plot(antithetic1, '--b')
line1, = plt.plot(regular1, '-y')

plt.xlabel('Paths')
plt.ylabel('Call Price')
plt.ylim(25, 50)

plt.legend([line0, line1], ['AntiThetic1', 'Regular1'], fontsize=14)
plt.tight_layout()
plt.show()
fig.savefig('trial1', dpi=1200)
```

```python
fig = plt.figure(figsize=(6,4))

line0, = plt.plot(antithetic2, '--b')
line1, = plt.plot(regular2, '-y')

plt.xlabel('Paths')
plt.ylabel('Call Price')
plt.ylim(60, 70)

plt.legend([line0, line1], ['AntiThetic2', 'Regular2'], fontsize=14)
plt.tight_layout()
plt.show()
fig.savefig('trial2', dpi=1200)
```
